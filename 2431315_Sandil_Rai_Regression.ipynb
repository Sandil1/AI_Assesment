Splitting the selectedDf into training and testing sets

```python
from sklearn.model_selection import train_test_split

X = selectedDf.drop(columns=['Life expectancy '])
Y = selectedDf['Life expectancy ']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

**Linear Regression Model From Scratch**

```python
# Feature Scaling (Standardization)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
sc_X_train = scaler.fit_transform(X_train)
sc_X_test = scaler.transform(X_test)
```

```python
# Initialize parameters
def initialize_params(d):
    b = 0  # Bias
    w = np.zeros(d)  # Weights
    return b, w
```

```python
# Linear function
def linear_function(b, w, X):
    return np.dot(X, w) + b
```

Cost Function along with L2(Ridge) Regularization

```python
# Cost function (MSE) along with ridges regularization
def cost_function(Y_true, Y_pred, w, lambda_):
    m = len(Y_true)
    mse = np.sum((Y_true - Y_pred) ** 2) / (2 * m)
    L2_regularization = (lambda_ / (2 * m)) * np.sum(w ** 2)
    return mse + L2_regularization
```

```python
# Gradient descent update
def update_parameters(Y_true, Y_pred, X, b, w, alpha, lambda_):
    m = len(Y_true)
    dw = -np.dot(X.T, (Y_true - Y_pred)) / m + (lambda_ / m) * w # L2 regularization
    db = -np.mean(Y_true - Y_pred)
    w -= alpha * dw
    b -= alpha * db
    return b, w
```

```python
# Training function
def train_linear_regression(X_train, Y_train, n_iter=10000, alpha=0.001, lambda_ = 1):
    b, w = initialize_params(X_train.shape[1])
    cost_history = []

    for i in range(n_iter):
        Y_hat = linear_function(b, w, X_train)
        cost_now = cost_function(Y_train, Y_hat, w, lambda_)
        b, w = update_parameters(Y_train, Y_hat, X_train, b, w, alpha, lambda_)
        cost_history.append(cost_now)

    return b, w, cost_history

# Train from-scratch model
b, w, cost_history = train_linear_regression(sc_X_train, Y_train, n_iter=10000, alpha=0.001)

# Predictions on test set
Y_test_pred_scratch = linear_function(b, w, sc_X_test)
```

**Evaluation Metrics: Linear Regression Model From Scratch**

**Mean-Squared Error (MSE)**

```python
from sklearn.metrics import mean_squared_error, r2_score

test_mse_scratch = mean_squared_error(Y_test, Y_test_pred_scratch)
train_mse_scratch = mean_squared_error(Y_train, linear_function(b, w, sc_X_train))

print(f"Mean-Squared Error during training (MSE): {train_mse_scratch:.4f}")
print(f"Mean-Squared Error during testing (MSE): {test_mse_scratch:.4f}")
```

**R-Squared Error**

```python
test_r2_scratch = r2_score(Y_test, Y_test_pred_scratch)
train_r2_scratch = r2_score(Y_train, linear_function(b, w, sc_X_train))

print(f"R-Squared Error during training (R2): {train_r2_scratch:.4f}")
print(f"R-Squared Error during testing (R2): {test_r2_scratch:.4f}")
```

**Decision Tree Model for Regression using sklearn library**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

decisionTreeReg = DecisionTreeRegressor(random_state = 42)
```

**Hyper-parameter Tuning: Decision Tree Model**

Parameter grid for hyperparameter tuning in Decision Tree model

```python
param_grid = {
    'max_depth': [3, 5, 7, 10],  # Depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],  # Minimum samples required at a leaf node
    'max_features': [None, 'sqrt', 'log2']  # Number of features to consider at each split
}
```

```python
grid_search = GridSearchCV(estimator = decisionTreeReg, param_grid = param_grid, cv = 3, scoring = 'neg_mean_squared_error', n_jobs = -1)
grid_search.fit(sc_X_train, Y_train)

print("Best Parameters:", grid_search.best_params_)
```

Decision Tree Model after hyper-parameter tuning

```python
best_dt_model = grid_search.best_estimator_
Y_test_pred_dt = best_dt_model.predict(sc_X_test)
```

```python
test_mse_dt = mean_squared_error(Y_test, Y_test_pred_dt)
train_mse_dt = mean_squared_error(Y_train, best_dt_model.predict(sc_X_train))

print(f"Mean-Squared Error during training (MSE): {train_mse_dt:.4f}")
print(f"Mean-Squared Error during testing (MSE): {test_mse_dt:.4f}")
```

```python
train_r2_dt = r2_score(Y_train, best_dt_model.predict(sc_X_train))
test_r2_dt = r2_score(Y_test, Y_test_pred_dt)

print(f"R-Squared Error during training (R2): {train_r2_dt:.4f}")
print(f"R-Squared Error during testing (R2): {test_r2_dt:.4f}")
```
